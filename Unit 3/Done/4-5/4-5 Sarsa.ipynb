```python
import time

#Make stopwatch
class Stopwatch:
    def __init__(self):
        self.start_time = None

    def start(self):
        self.start_time = time.time()

    def elapsed_time(self):
        if self.start_time is None:
            return 0
        else:
            return time.time() - self.start_time
stopwatch = Stopwatch()
```
